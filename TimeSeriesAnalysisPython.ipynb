```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy import stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
%matplotlib inline
```

```python
global_temp=pd.read_csv('GlobalTemperatures.csv', index_col='dt', parse_dates=True)
```

```python
list(global_temp)
```

<h3> Time Series Analysis => Land Average Temperature

```python
plt.figure(figsize=(20,5))
global_temp['LandAverageTemperature'].plot(color='blue')
plt.xlabel('Year')
plt.ylabel('Land Average Temperature')
plt.grid()
plt.show()
```

<h4> Resampling data in column named LandAverageTemperature to analyze better </h4>

```python
sample_A=global_temp.resample('A').mean()
plt.subplot(4,1,3)
sample_A['LandAverageTemperature'].plot(color='blue', figsize=(10,10))
plt.ylim([0,18])
plt.xlabel("Year")
plt.grid()
```

<h4> It can be inferred from the graph that the average temperature on land is slightly above the mean over the past few years. The mean being 8.386 with max and min value being 19.021 and -2.080 respectively as described below </h4>

```python
global_temp["LandAverageTemperature"].describe()
```

```python
plt.figure(figsize=(15,4)) 
global_temp['LandAndOceanAverageTemperature'].plot(color='blue') 
plt.xlabel('Year') 
plt.ylabel('Land and Ocean Average Temperature') 
plt.grid() 
plt.show() 
plt.figure(figsize=(15,16))
```

<h3> 10 Year Moving Average </h2>

<h4> The code below is to turn the whole data into a 10 year moving average analysis.
Here, n=120 which is nothing but 12x10 for 10 years rolling average </h4>

```python
LandAndOceanAverageTemperature = global_temp['LandAndOceanAverageTemperature'] 
LandAndOceanAverageTemperature.rolling(120).mean().plot(figsize=(25,14), linewidth=5, fontsize=15, label='10-Year Rolling Average',color='red') 
plt.legend(loc=1,fontsize=15) 
plt.xlabel('Year', fontsize=15) 
plt.grid() 
plt.ylabel('Land and Ocean Average Temperature', fontsize=15) 
plt.title('10-Year Rolling Average of Land and Ocean Average Temperature', fontsize=20)
```

<h4> A moving average is used for analysis when a smoother representation of data is required with short-term visibility in fluctuations and it is capable of highlighting trends in long-term. The above graph shows moving average of LandAndOceanAverageTemperature from 1860 to 2015.
    <p>
    
The trend in long-term is perceived to be upwards which clearly indicates that it is on the rise from year 1860 to 2015. However, this rate of rise seems to be at faster pace after 1980s. </h4>
