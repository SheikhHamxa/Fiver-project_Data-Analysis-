```python
# removing non useful columns and splitting as well
Regular_blood_test=['Proteina C reativa', 'Neutrophils', 'Mean platelet volume', 'Monocytes', 'Red blood cell distribution width', 'Red blood Cells', 'Platelets', 'Eosinophils', 'Basophils',
                    'Leukocytes', 'Mean corpuscular hemoglobin', 'Mean corpuscular volume', 'Mean corpuscular hemoglobin concentration', 'Lymphocytes', 'Hemoglobin', 'Hematocrit']
Influenza_rapid_test=['Influenza B rapid test', 'Influenza A rapid test']
Target_variable=['SARS-Cov-2 exam result']
Other_to_keep=['Patient age quantile']

df=df[Target_variable+Other_to_keep+Regular_blood_test+Influenza_rapid_test]
df.shape
```

```python
df
```

```python
# convert to bolean values as this will be used as predicting column
convert={'negative': 0, 'positive': 1}
df.replace(convert,inplace=True)
```

```python
df
```

```python
# DATA CLEANING
# Remove cases with no regualr blood test results and impute remaining incomplete test results using mean
df=df[df[Regular_blood_test].count(axis=1)!=0] 
for col in Regular_blood_test:
    df[col].fillna(df[col].mean(), inplace=True)
```

```python
df=df[df[ Influenza_rapid_test].count(axis=1)!=0] 

for col in Influenza_rapid_test:
    df[col].fillna(df[col].mean(), inplace=True)
```

```python
df
```

```python
x=df.drop(['SARS-Cov-2 exam result'],axis=1)
y=df['SARS-Cov-2 exam result']
```

```python
x.shape
```

```python
y.shape
```

```python
# covert to numpy matrix
# to get dummy values
x=pd.get_dummies(x)
x_mat = x.to_numpy()
```

```python
# setting random state
from sklearn.model_selection import train_test_split
rs = 10
#split the test/train data
x_train, x_test, y_train, y_test = train_test_split(x_mat, y, test_size=0.3, stratify=y, random_state=rs)
```

```python
#STANDARDISATION REQUIRED FOR REGRESSION MODELS
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()# initialise a standard scaler object

x_train = scaler.fit_transform(x_train, y_train) #find transformation from training data and use it to standardise
x_test = scaler.transform(x_test) #standardise test data    
```

```python
from sklearn.linear_model import LogisticRegression

model_reg = LogisticRegression(random_state=rs)

# fit it to training data
model_reg.fit(x_train, y_train)
```

```python
# grid search CV
from sklearn.model_selection import train_test_split, GridSearchCV

params = {'C': [pow(10, x) for x in range(-6, 4)]}

# use all cores to tune logistic regression with C parameter
reg_cv = GridSearchCV(param_grid=params, estimator=LogisticRegression(random_state=rs),return_train_score=True, cv=10, n_jobs=-1)
reg_cv.fit(x_train, y_train)
print(reg_cv.best_params_)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
result_set = reg_cv.cv_results_

train_result = result_set['mean_train_score']
test_result = result_set['mean_test_score']
print("Total number of models: ", len(test_result))
# plot Hyperparameter C values vs training and test accuracy score
plt.plot(range(0, len(train_result)), train_result, 'b', range(0,len(test_result)), test_result, 'r')
plt.xlabel('Hyperparameter C\nBlue = training acc. Red = test acc.')
plt.xticks(range(0, len(train_result)),[pow(10, x) for x in range(-6, 4)])
plt.ylabel('score')
plt.show()
```

```python
# training and test accuracy
print('(F)')
print('C=1, not CV')
print("Train accuracy default:", model_reg.score(x_train, y_train))
print("Test accuracy default:", model_reg.score(x_test, y_test))
print('C=0.01, CV')
print("Train accuracy GridSearchCV:", reg_cv.score(x_train, y_train)) #Grid search
print("Test accuracy GridSearchCV:", reg_cv.score(x_test, y_test))
```

```python
feature_names = x.columns
coef = model_reg.coef_[0]

# sort them out in descending order
indices = np.argsort(np.absolute(coef))
indices = np.flip(indices, axis=0)

print('(E)')
print('The 5 most important variables are:')
indices = indices[:5]# limit to top5
for i in indices:
    print(feature_names[i], ':', coef[i])
    
print('(F)')

```

### 3.2 Recursive feature elimination + CV grid search regression


```python
import sklearn.model_selection
sklearn.model_selection.KFold
from sklearn.feature_selection import RFECV
logreg_model = LogisticRegression()
rfecv = RFECV(estimator=logreg_model, step=1, scoring='accuracy')

rfecv.fit(x, y)
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.title('Logistic Regression CV score vs No of Features')
plt.xlabel("Number of features selected")
plt.ylabel("Cross validation score (nb of correct classifications)")
plt.plot(range(1, len(rfecv.grid_scores_) + 1), rfecv.grid_scores_)
plt.show()
```

```python
feature_importance = list(zip(feature_names, rfecv.support_))
new_features = []
for key,value in enumerate(feature_importance):
    if(value[1]) == True:
        new_features.append(value[0])
        
print(new_features)
```

```python
from sklearn.model_selection import cross_val_score

X_new = df[new_features]
initial_score = cross_val_score(logreg_model, x, y, scoring='accuracy').mean()
print("Initial accuracy : {} ".format(initial_score))
fe_score = cross_val_score(logreg_model, X_new, y,scoring='accuracy').mean()
print("Accuracy after Feature Selection : {} ".format(fe_score))
```

# the Receiver Operating Characteristic (ROC) curve.

<p> Receiver operating characteristic (ROC) Curve
Receiver Operating Characteristic (ROC) curve is a plot of the true positive rate against the false positive rate. The AUC is the Area Under Curve. If the AUC is high, the model is better distinguishing between positive and negative class. The ROC curve is plotted with “True Positive Rate” or Recall (on the y-axis) against the “False Positive Rate” (on the x-axis). When the AUC is 0.5 means that the model has no discrimination capacity to distinguish between positive and negative class.

. </p>
